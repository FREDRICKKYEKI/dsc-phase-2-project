```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data =pd.read_csv('data\kc_house_data.csv')
data.head()
```

```python
data.info()
```

```python
#checking missing values
data.isnull().sum()
```

```python
#we shall drop the few missing values in view column
data.dropna(subset=['view'],inplace=True)
data.head()
```

```python
#check if the values have been dropped
data.view.isnull().sum()
```

```python
# Fill missing values in 'waterfront' with 0 and in 'yr_renovated' with the median
fill_values = {'waterfront': 0, 'yr_renovated': data['yr_renovated'].median()}
data.fillna(fill_values, inplace=True)
```

```python
#checking missing values
data.isnull().sum()
```

```python
data.describe
```

```python
#checking for duplicates
data.duplicated().sum()
```

```python
data
```

```python
# Dropping irrelevant columns 
columns_to_drop = ['id', 'date','zipcode','lat','long','sqft_living15','sqft_lot15']
data.drop(columns=columns_to_drop, inplace=True) 
```

```python
data.head()
```

```python
data.info()
```

```python
Q1 = data['price'].quantile(0.25)
Q3 = data['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['price'] < lower_bound) | (data['price'] > upper_bound)]
```

```python
outliers
```

```python
#Apply a log transformation to reduce the impact of outliers and make the distribution more normal.
data['Log_price'] = np.log(data['price'])
```
